## Plot

I plotted a **regression of win percentage on sugar percentage**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load the data
file_path = "https://raw.githubusercontent.com/fivethirtyeight/data/master/candy-power-ranking/candy-data.csv"
df = pd.read_csv(file_path)

# Select relevant columns
x_col = "sugarpercent"
y_col = "winpercent"

# Drop missing values if any
df = df[[x_col, y_col]].dropna()

# Perform linear regression
X = sm.add_constant(df[x_col])  # Add constant for intercept
model = sm.OLS(df[y_col], X).fit()

# Plot the regression
plt.figure(figsize=(8, 5))
sns.regplot(x=df[x_col], y=df[y_col], ci=None, line_kws={"color": "red"})
plt.xlabel("Sugar Percentage")
plt.ylabel("Win Percentage")
plt.title("Regression of Win Percentage on Sugar Percentage")

# Show plot
plt.show()

# Display regression summary
print(model.summary())
```
